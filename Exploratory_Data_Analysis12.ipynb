```python
# ============================
# 4. LOAD DATA
# ============================
path = "/content/february_sales.csv"

df = pd.read_csv(path)

df.head()
```

```python
# ============================
# 5. DATA QUALITY CHECKS
# ============================
print("\n============================")
print(" DATA QUALITY CHECKS")
print("============================")

print("\nâž¡ Dataset shape:", df.shape)
print("\nâž¡ Data types:")
print(df.dtypes)

print("\nâž¡ Missing values:")
print(df.isnull().sum())

print("\nâž¡ Duplicate rows:", df.duplicated().sum(), "\n")

# Summary statistics
df.describe(include='all')
```

```python
# ============================
# 6. MISSING DATA VISUALIZATION
# ============================
plt.figure(figsize=(10,4))
msno.matrix(df)
plt.show()

plt.figure(figsize=(6,4))
msno.bar(df)
plt.show()
```

```python
# ============================
# 7. DATA AUDIT REPORT
# ============================
numeric_cols = df.select_dtypes(include=['number']).columns

audit = pd.DataFrame({
    "Column": df.columns,
    "Data Type": df.dtypes.values,
    "Missing Count": df.isnull().sum().values,
    "Unique Values": df.nunique().values,
})

print("\nDATA AUDIT REPORT")
audit
```

```python
# ============================
# 8. BASIC CLEANING
# ============================
df = df.drop_duplicates()

# Fill missing based on type
for col in df.columns:
    if df[col].dtype == "object":
        df[col] = df[col].fillna("Unknown")
    else:
        df[col] = df[col].fillna(df[col].median())

print(" Cleaning complete")
```

```python
# ============================
# 9. EXPLORATORY VISUALIZATIONS
# ============================

print("\nðŸ“Š BASIC VISUALIZATIONS")

# Sales trend line (only if sales + date exist)
if "Date" in df.columns and "Sales" in df.columns:
    df["Date"] = pd.to_datetime(df["Date"], errors='coerce')
    df = df.sort_values("Date")

    plt.figure(figsize=(12,5))
    sns.lineplot(data=df, x="Date", y="Sales")
    plt.title("Daily February Sales Trend")
    plt.xticks(rotation=45)
    plt.show()

# Histogram for all numeric variables
df.hist(figsize=(14,8), bins=20)
plt.suptitle("Distribution of Numeric Variables")
plt.show()

# Correlation heatmap (only if numeric exists)
if len(numeric_cols) > 1:
    plt.figure(figsize=(10,6))
    sns.heatmap(df[numeric_cols].corr(), annot=True, cmap="coolwarm")
    plt.title("Correlation Matrix")
    plt.show()
```

```python
# ============================
# 10. ADVANCED VISUALIZATIONS
# ============================

print("\nADVANCED VISUALIZATIONS")

# Pairplot
if len(numeric_cols) >= 2:
    sns.pairplot(df[numeric_cols])
    plt.show()

# Boxplots
plt.figure(figsize=(12,5))
sns.boxplot(data=df[numeric_cols])
plt.title("Outlier Detection")
plt.xticks(rotation=45)
plt.show()

# Treemap (if product exists)
if "Product" in df.columns and "Sales" in df.columns:
    fig = px.treemap(df, path=["Product"], values="Sales",
                     title="Product Sales Treemap")
    fig.show()
```

```python
# ============================
# 11. PCA (HIDDEN STRUCTURE)
# ============================

print("\n PCA â€“ HIDDEN STRUCTURE")

# PCA only if 2+ numeric cols
if len(numeric_cols) >= 2:
    pca = PCA(n_components=2)
    pca_data = pca.fit_transform(df[numeric_cols])

    plt.figure(figsize=(8,6))
    plt.scatter(pca_data[:,0], pca_data[:,1])
    plt.title("PCA 2D Projection (Hidden Structure)")
    plt.xlabel("PC1")
    plt.ylabel("PC2")
    plt.show()
else:
    print("Not enough numeric columns for PCA.")
```

```python
# ============================
# 12. AUTOMATED VISUALIZATION GENERATOR
# ============================

def auto_visualize(df):
    numeric_cols = df.select_dtypes(include=['number']).columns

    for col in numeric_cols:
        plt.figure(figsize=(7,4))
        sns.histplot(df[col], kde=True)
        plt.title(f"Auto-Plot: Distribution of {col}")
        plt.show()

print("\n Running automated visualization...")
auto_visualize(df)
```

```python
# ============================
# 13. FEATURE RELATIONSHIPS
# ============================

print("\n FEATURE RELATIONSHIPS TO SALES")

if "Sales" in df.columns:
    for col in numeric_cols:
        if col != "Sales":
            plt.figure(figsize=(6,4))
            sns.regplot(data=df, x=col, y="Sales")
            plt.title(f"{col} vs Sales")
            plt.show()
```
