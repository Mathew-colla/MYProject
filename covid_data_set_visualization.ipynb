####Covid Data Set visualization

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the dataset
file_path= '/content/owid-covid-data.csv'
df = pd.read_csv(file_path)

# Convert 'date' column to datetime objects
df['date'] = pd.to_datetime(df['date'])

# Display the first few rows and basic info
print("DataFrame head:")
print(df.head())
print("\nDataFrame info:")
df.info()

# Example visualization: New cases over time for a specific country
country = 'United States'
country_df = df[df['location'] == country].sort_values('date')

plt.figure(figsize=(12, 6))
sns.lineplot(x='date', y='new_cases', data=country_df)
plt.title(f'New COVID-19 Cases in {country} Over Time')
plt.xlabel('Date')
plt.ylabel('New Cases')
plt.grid(True)
plt.show()
```
